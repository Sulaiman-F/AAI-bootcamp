Key Differences Between Histograms and Bar Plots

**Data Type:** Histograms are for continuous data (e.g., age, temperature, test scores). Bar plots are for categorical data (e.g., types of cars, names, countries).


**Spacing:** Histograms have no gaps between bars because the data is continuous. Bar plots have distinct spaces between bars.

**Order:** Histogram bars must be in numerical order, representing a range of values. Bar plot bars can be ordered (e.g., largest to smallest) or unordered without losing meaning.

**Purpose:** Histograms are used to analyze data distribution, skewness, and frequency. Bar plots are designed to compare the size of different categories.

## B. Subplots and Grids

**1.** Using a seaborn **figure-level** function (like `sns.relplot` or `sns.displot`), create subplots using the `col` and `row` parameters. For example, plot `flipper_length_mm` vs `body_mass_g` across columns for `sex` and rows for `island`.

```python
sns.relplot(
    data=df, 
    x="bill_length_mm", 
    y="bill_depth_mm", 
    row="island", col="sex")
```

**2.** Using **axes-level** functions, create a figure with subplots (e.g., 1 row and 2 columns) using `plt.subplots`. 
- Plot a scatterplot of `bill_length_mm` vs `bill_depth_mm` on the first axes (`ax1`).
- Plot a boxplot (or another axes-level plot of your choice) of `bill_length_mm` by `species` on the second axes (`ax2`).
- Make sure to use `sharey=True` (or `sharex=True`) where appropriate so the subplots are linked on the shared dimension.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5), sharey=True)

sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", ax=ax1)
ax1.set_title("Scatter: Bill Length vs Depth")

sns.boxplot(data=df, x="species", y="bill_length_mm", ax=ax2)
ax2.set_title("Boxplot: Bill Length by Species")

plt.tight_layout()
plt.show()
```
