## Task 5: Confidence Scores
**Confidence grades** were introduced in [v2.34.0](https://github.com/docling-project/docling/releases/tag/v2.34.0) to help users understand how well a conversion performed and guide decisions about post-processing workflows. They are available in the [`confidence`](https://docling-project.github.io/docling/reference/document_converter/#docling.document_converter.ConversionResult.confidence) field of the [`ConversionResult`](https://docling-project.github.io/docling/reference/document_converter/#docling.document_converter.ConversionResult) object returned by the document converter.

Your task is to look for the [Confidence Grades](https://docling-project.github.io/docling/concepts/confidence_scores/#scores-and-grades) and report them.
